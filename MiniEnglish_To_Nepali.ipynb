
***Problem***

    -Transformers cannot directly work with text, so they requeire numbers as input. 
    - How to ensure all sentence batch have same length.
    - How to make start and end of sentence in decoder.
    

***Solution***

    - Covert each word  to ID(tokenization $ vocabulary)
    - Use Padding to ensure same length
    - Add special tokens :
        <pad> → 0 (for padding)
        <sos> → 1 (start of sentence)
        <eos> → 2 (end of sentence)
                

```python
# 2(a) Vocabulary Creation

# Special tokens
PAD_TOKEN = "<pad>"
SOS_TOKEN = "<sos>"
EOS_TOKEN = "<eos>"

# Function to create vocabulary from a list of sentences
def build_vocab(sentences):
    vocab = {PAD_TOKEN: 0, SOS_TOKEN: 1, EOS_TOKEN: 2}  # start with special tokens
    idx = 3  # start indexing words from 3
    for sentence in sentences:
        for word in sentence.split():  # simple split by space
            if word not in vocab:
                vocab[word] = idx
                idx += 1
    return vocab

# Create vocabularies
src_vocab = build_vocab(english_sentences)  # English
tgt_vocab = build_vocab(nepali_sentences)   # Nepali

# Inverse vocab for decoding later (ID -> word)
inv_src_vocab = {i: w for w, i in src_vocab.items()}
inv_tgt_vocab = {i: w for w, i in tgt_vocab.items()}

# Print some info
print("English vocab size:", len(src_vocab))
print("Nepali vocab size:", len(tgt_vocab))
print("\nSample English vocab:", list(src_vocab.items())[:10])
print("\nSample Nepali vocab:", list(tgt_vocab.items())[:10])
```

```python
# 2b.  Tokenize and Pad sentence
import torch
from torch.nn.utils.rnn import pad_sequence

# Function to encode a sentence into token IDs
def encode_sentence(sentence, vocab, add_sos_eos=False):
    tokens = [vocab[word] for word in sentence.split()]
    if add_sos_eos:
        tokens = [vocab[SOS_TOKEN]] + tokens + [vocab[EOS_TOKEN]]
    return torch.tensor(tokens, dtype=torch.long)

# Encode all English sentences (source)
src_encoded = [encode_sentence(s, src_vocab) for s in english_sentences]

# Encode all Nepali sentences (target) with <sos> and <eos>
tgt_encoded = [encode_sentence(s, tgt_vocab, add_sos_eos=True) for s in nepali_sentences]

# Pad sequences so all sentences have same length
src_padded = pad_sequence(src_encoded, batch_first=True, padding_value=src_vocab[PAD_TOKEN])
tgt_padded = pad_sequence(tgt_encoded, batch_first=True, padding_value=tgt_vocab[PAD_TOKEN])

# Print shapes and some examples
print("Source padded shape:", src_padded.shape)
print("Target padded shape:", tgt_padded.shape)

print("\nFirst English sentence IDs:", src_padded[0])
print("First Nepali sentence IDs:", tgt_padded[0])
```

***Problem***

    -Transformers cannot directly pass token ID so we need vectors that repesent word meaning. Thus,
    -Convert token IDs to Embedding vectors

    -Transformers also need to know word order, because attention alone has no sense of sequence. Thus,
    -Add positional encoding.
    

***Solution***

    - Use PyTorch nn.Embedding for source (English) sentences.
    -Compute positional encodings using sinusoidal formulas.
    -Add positional encodings to embeddings.
    Output tensor → [batch_size, seq_len, emb_dim], ready for encoder layers.

```python
# Step 3a: Create Embeddings

import math
import torch.nn as nn

# Parameters
EMB_SIZE = 32  # embedding dimension
BATCH_SIZE = src_padded.shape[0]
SEQ_LEN = src_padded.shape[1]

# Source embedding
src_emb = nn.Embedding(len(src_vocab), EMB_SIZE)

# Get embedded input
embedded_input = src_emb(src_padded)  # shape: [batch_size, seq_len, emb_dim]

print("Embedded input shape:", embedded_input.shape)
print("First word embedding of first sentence:\n", embedded_input[0,0])
```

```python
# Positional encoding

# Create positional encoding matrix
pos_encoding = torch.zeros(SEQ_LEN, EMB_SIZE)
for pos in range(SEQ_LEN):
    for i in range(0, EMB_SIZE, 2):  # even dimensions
        pos_encoding[pos, i] = math.sin(pos / (10000 ** (i / EMB_SIZE)))
        if i+1 < EMB_SIZE:
            pos_encoding[pos, i+1] = math.cos(pos / (10000 ** (i / EMB_SIZE)))

# Expand to batch size
pos_encoding = pos_encoding.unsqueeze(0).repeat(BATCH_SIZE, 1, 1)

# Add positional encoding to embeddings
embedded_with_pos = embedded_input + pos_encoding

print("Embedded input with positional encoding shape:", embedded_with_pos.shape)
print("First word embedding with positional encoding for first sentence:\n", embedded_with_pos[0,0])
```

4. Multi-Head Self-Attention in the Eoncoder

***Problem***
    
    - Words in a sentence are not independent.
     
    -Each word should attend to all other words to understand context
     So, we need 
         i. A mechanism to let each word see other words in the sentence → Self-Attention.
         ii.Multiple “attention heads” to capture different types of relationships → Multi-Head Attention.

***Solution***

    - Apply nn.MultiheadAttention. 
    - With different Query, Key and Values


```python
#4a Multihead Attention
# Parameters
NHEAD = 4  # number of attention heads

# PyTorch MultiheadAttention
self_attn = nn.MultiheadAttention(embed_dim=EMB_SIZE, num_heads=NHEAD, batch_first=True)

# embedded_with_pos shape: [batch_size, seq_len, emb_dim]
# For self-attention, query=key=value=embedded input
attn_output, attn_weights = self_attn(embedded_with_pos, embedded_with_pos, embedded_with_pos)

print("Attention output shape:", attn_output.shape)  # [batch_size, seq_len, emb_dim]
print("Attention weights shape:", attn_weights.shape)  # [batch_size, seq_len, seq_len]
print("\nAttention weights for first word in first sentence:\n", attn_weights[0,0])
```

```python
# Encoder layer
# Parameters
FFN_HID_DIM = 64  # hidden dimension in feed-forward
NUM_LAYERS = 2    # number of encoder layers

# Define one encoder layer
class EncoderLayer(nn.Module):
    def __init__(self, emb_size, nhead, ffn_hid_dim, dropout=0.1):
        super().__init__()
        self.self_attn = nn.MultiheadAttention(embed_dim=emb_size, num_heads=nhead, batch_first=True)
        self.norm1 = nn.LayerNorm(emb_size)
        self.norm2 = nn.LayerNorm(emb_size)
        self.ffn = nn.Sequential(
            nn.Linear(emb_size, ffn_hid_dim),
            nn.ReLU(),
            nn.Linear(ffn_hid_dim, emb_size)
        )
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, x):
        # Multi-head self-attention
        attn_output, attn_weights = self.self_attn(x, x, x)
        x = self.norm1(x + self.dropout(attn_output))  # Residual + LayerNorm
        
        # Feed-Forward Network
        ffn_output = self.ffn(x)
        x = self.norm2(x + self.dropout(ffn_output))  # Residual + LayerNorm
        return x

# Stack multiple encoder layers
class Encoder(nn.Module):
    def __init__(self, num_layers, emb_size, nhead, ffn_hid_dim):
        super().__init__()
        self.layers = nn.ModuleList([
            EncoderLayer(emb_size, nhead, ffn_hid_dim) for _ in range(num_layers)
        ])
        
    def forward(self, x):
        for layer in self.layers:
            x = layer(x)
        return x

# Instantiate encoder
encoder = Encoder(NUM_LAYERS, EMB_SIZE, NHEAD, FFN_HID_DIM)

# Forward pass through encoder
encoder_output = encoder(embedded_with_pos)

print("Encoder output shape:", encoder_output.shape)  # [batch_size, seq_len, emb_dim]
print("First word representation of first sentence:\n", encoder_output[0,0])
```

```python
# decoder

class DecoderLayer(nn.Module):
    def __init__(self, emb_size, nhead, ffn_hid_dim, dropout=0.1):
        super().__init__()
        # Masked self-attention for target sequence
        self.self_attn = nn.MultiheadAttention(embed_dim=emb_size, num_heads=nhead, batch_first=True)
        # Attention over encoder output
        self.enc_attn = nn.MultiheadAttention(embed_dim=emb_size, num_heads=nhead, batch_first=True)
        # Feed-forward network
        self.ffn = nn.Sequential(
            nn.Linear(emb_size, ffn_hid_dim),
            nn.ReLU(),
            nn.Linear(ffn_hid_dim, emb_size)
        )
        # Layer norms
        self.norm1 = nn.LayerNorm(emb_size)
        self.norm2 = nn.LayerNorm(emb_size)
        self.norm3 = nn.LayerNorm(emb_size)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x, enc_out, tgt_mask=None):
        # Masked self-attention
        attn_output, _ = self.self_attn(x, x, x, attn_mask=tgt_mask)
        x = self.norm1(x + self.dropout(attn_output))
        
        # Encoder-decoder attention
        attn_output, _ = self.enc_attn(x, enc_out, enc_out)
        x = self.norm2(x + self.dropout(attn_output))
        
        # Feed-forward
        ffn_output = self.ffn(x)
        x = self.norm3(x + self.dropout(ffn_output))
        
        return x

# Decoder stack
class Decoder(nn.Module):
    def __init__(self, num_layers, emb_size, nhead, ffn_hid_dim):
        super().__init__()
        self.layers = nn.ModuleList([DecoderLayer(emb_size, nhead, ffn_hid_dim) for _ in range(num_layers)])
    
    def forward(self, x, enc_out, tgt_mask=None):
        for layer in self.layers:
            x = layer(x, enc_out, tgt_mask)
        return x
```

```python
# Look_ Ahead Mask(Decoder)
# Function to create look-ahead mask
def create_look_ahead_mask(size):
    mask = torch.triu(torch.ones(size, size), diagonal=1).bool()  # upper triangular
    return mask
```

```python
class TransformerModel(nn.Module):
    def __init__(self, src_vocab_size, tgt_vocab_size, emb_size, nhead, ffn_hid_dim, num_layers):
        super().__init__()
        # Embeddings
        self.src_emb = nn.Embedding(src_vocab_size, emb_size)
        self.tgt_emb = nn.Embedding(tgt_vocab_size, emb_size)
        
        # Positional encoding
        self.pos_encoding = self.create_positional_encoding(100, emb_size)  # max seq len 100
        
        # Encoder and decoder
        self.encoder = Encoder(num_layers, emb_size, nhead, ffn_hid_dim)
        self.decoder = Decoder(num_layers, emb_size, nhead, ffn_hid_dim)
        
        # Output linear layer
        self.fc_out = nn.Linear(emb_size, tgt_vocab_size)
    
    def create_positional_encoding(self, max_len, emb_size):
        pos_encoding = torch.zeros(max_len, emb_size)
        for pos in range(max_len):
            for i in range(0, emb_size, 2):
                pos_encoding[pos, i] = math.sin(pos / (10000 ** (i / emb_size)))
                if i+1 < emb_size:
                    pos_encoding[pos, i+1] = math.cos(pos / (10000 ** (i / emb_size)))
        return pos_encoding.unsqueeze(0)  # [1, max_len, emb_size]

    def forward(self, src, tgt):
        batch_size, src_len = src.shape
        batch_size, tgt_len = tgt.shape
        
        # Encoder
        src_emb = self.src_emb(src) + self.pos_encoding[:, :src_len, :]
        enc_out = self.encoder(src_emb)
        
        # Decoder
        tgt_emb = self.tgt_emb(tgt) + self.pos_encoding[:, :tgt_len, :]
        look_ahead_mask = create_look_ahead_mask(tgt_len).to(tgt.device)
        dec_out = self.decoder(tgt_emb, enc_out, tgt_mask=look_ahead_mask)
        
        # Project to vocabulary
        out = self.fc_out(dec_out)  # [batch, tgt_len, tgt_vocab_size]
        return out
```

```python
SRC_VOCAB_SIZE = len(src_vocab)
TGT_VOCAB_SIZE = len(tgt_vocab)

model = TransformerModel(SRC_VOCAB_SIZE, TGT_VOCAB_SIZE, EMB_SIZE, NHEAD, FFN_HID_DIM, NUM_LAYERS)

# Test forward pass
sample_src = src_padded[:2]  # first 2 sentences
sample_tgt = tgt_padded[:2, :-1]  # input to decoder (without last token)

output = model(sample_src, sample_tgt)
print("Transformer output shape:", output.shape)  # [batch, tgt_len, tgt_vocab_size]
```

```python
# Preparring for training
import torch.nn as nn
import torch.optim as optim

# Target input and expected output
tgt_input = tgt_padded[:, :-1]  # all tokens except last (<eos>)
tgt_output = tgt_padded[:, 1:]  # all tokens except first (<sos>)

# Define loss and optimizer
criterion = nn.CrossEntropyLoss(ignore_index=tgt_vocab[PAD_TOKEN])
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
EPOCHS = 500

for epoch in range(EPOCHS):
    model.train()
    optimizer.zero_grad()
    
    # Forward pass
    output = model(src_padded, tgt_input)  # [batch, tgt_len, vocab_size]
    
    # Reshape for loss: [batch * tgt_len, vocab_size] vs [batch * tgt_len]
    loss = criterion(output.reshape(-1, TGT_VOCAB_SIZE), tgt_output.reshape(-1))
    
    # Backpropagation
    loss.backward()
    optimizer.step()
    
    if (epoch+1) % 10 == 0:
        print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}")

print("Training finished!")
```

```python
def translate(model, sentence, src_vocab, tgt_vocab, inv_tgt_vocab, max_len=20):
    model.eval()
    
    # Encode source sentence
    src_tokens = torch.tensor([src_vocab[word] for word in sentence.split()]).unsqueeze(0)  # [1, seq_len]
    
    # Initialize target sequence with <sos>
    tgt_tokens = torch.tensor([[tgt_vocab[SOS_TOKEN]]], dtype=torch.long)
    
    output_sentence = []
    
    for _ in range(max_len):
        # Forward pass
        out = model(src_tokens, tgt_tokens)
        
        # Get last token probabilities
        next_token = out[0, -1, :].argmax(-1).item()
        
        # Stop if <eos> is predicted
        if next_token == tgt_vocab[EOS_TOKEN]:
            break
        
        # Append word to output
        output_sentence.append(inv_tgt_vocab[next_token])
        
        # Append token to decoder input
        tgt_tokens = torch.cat([tgt_tokens, torch.tensor([[next_token]])], dim=1)
    
    return " ".join(output_sentence)
```

```python
# Example translation
english_test = "I am happy"
nepali_translation = translate(model, english_test, src_vocab, tgt_vocab, inv_tgt_vocab)
print("English:", english_test)
print("Predicted Nepali:", nepali_translation)
```
